```python
## Q5 Best Customer
query = """
SELECT c.customer_id,
       c.first_name,
       c.last_name,
       SUM(i.total) AS total_spent
FROM customers c
JOIN invoices i
ON c.customer_id = i.customer_id
GROUP BY c.customer_id, c.first_name, c.last_name
ORDER BY total_spent DESC
LIMIT 1;
"""

pd.read_sql(query, conn)
```

```python
## Q6 Rock Music Listeners 
query = """
SELECT DISTINCT c.email,
       c.first_name,
       c.last_name,
       g.name AS genre
FROM customers c
JOIN invoices i
ON c.customer_id = i.customer_id
JOIN invoice_items ii
ON i.invoice_id = ii.invoice_id
JOIN tracks t
ON ii.track_id = t.track_id
JOIN genres g
ON t.genre_id = g.genre_id
WHERE g.name = 'Rock'
ORDER BY c.email;
"""

pd.read_sql(query, conn)
```

```python
## Q7 Top 10 Rock Bands 
query = """
SELECT ar.name AS artist_name,
       COUNT(t.track_id) AS rock_track_count
FROM artists ar
JOIN albums al
ON ar.artist_id = al.artist_id
JOIN tracks t
ON al.album_id = t.album_id
JOIN genres g
ON t.genre_id = g.genre_id
WHERE g.name = 'Rock'
GROUP BY ar.name
ORDER BY rock_track_count DESC
LIMIT 10;
"""

pd.read_sql(query, conn)
```

```python
## Q8 Tracks Longer Than Average Length 
query = """
SELECT name, milliseconds
FROM tracks
WHERE milliseconds > (
        SELECT AVG(milliseconds)
        FROM tracks
)
ORDER BY milliseconds DESC;
"""

pd.read_sql(query, conn)
```

```python
## Q9 Amount Spent by Each Customer on Artists 
query = """
SELECT 
    c.first_name || ' ' || c.last_name AS customer_name,
    ar.name AS artist_name,
    SUM(ii.unit_price * ii.quantity) AS total_spent
FROM customers c
JOIN invoices i
    ON c.customer_id = i.customer_id
JOIN invoice_items ii
    ON i.invoice_id = ii.invoice_id
JOIN tracks t
    ON ii.track_id = t.track_id
JOIN albums al
    ON t.album_id = al.album_id
JOIN artists ar
    ON al.artist_id = ar.artist_id
GROUP BY customer_name, ar.name
ORDER BY total_spent DESC;
"""

pd.read_sql(query, conn)
```

```python
## Q10 Most Popular Genre for Each Country 
query = """
WITH genre_purchases AS (
    SELECT 
        c.country,
        g.name AS genre,
        COUNT(ii.quantity) AS purchases
    FROM customers c
    JOIN invoices i 
        ON c.customer_id = i.customer_id
    JOIN invoice_items ii 
        ON i.invoice_id = ii.invoice_id
    JOIN tracks t 
        ON ii.track_id = t.track_id
    JOIN genres g 
        ON t.genre_id = g.genre_id
    GROUP BY c.country, g.name
),
ranked_genres AS (
    SELECT *,
           RANK() OVER(PARTITION BY country ORDER BY purchases DESC) AS rank
    FROM genre_purchases
)
SELECT country, genre, purchases
FROM ranked_genres
WHERE rank = 1
ORDER BY country;
"""

pd.read_sql(query, conn)
```

```python
## Q11 Top Customer in Each Country 
query = """
WITH customer_spending AS (
    SELECT 
        c.country,
        c.first_name || ' ' || c.last_name AS customer_name,
        SUM(i.total) AS total_spent
    FROM customers c
    JOIN invoices i
        ON c.customer_id = i.customer_id
    GROUP BY c.country, customer_name
),
ranked_customers AS (
    SELECT *,
           RANK() OVER(PARTITION BY country ORDER BY total_spent DESC) AS rank
    FROM customer_spending
)
SELECT country, customer_name, total_spent
FROM ranked_customers
WHERE rank = 1
ORDER BY country;
"""

pd.read_sql(query, conn)
```

```python
## Q12 Most Popular Artists 
query = """
SELECT 
    ar.name AS artist_name,
    COUNT(ii.quantity) AS total_purchases
FROM artists ar
JOIN albums al
    ON ar.artist_id = al.artist_id
JOIN tracks t
    ON al.album_id = t.album_id
JOIN invoice_items ii
    ON t.track_id = ii.track_id
GROUP BY ar.name
ORDER BY total_purchases DESC
LIMIT 10;
"""

pd.read_sql(query, conn)
```

```python
## Q13 Most Popular Song 
query = """
SELECT 
    t.name AS track_name,
    COUNT(ii.quantity) AS purchase_count
FROM tracks t
JOIN invoice_items ii
    ON t.track_id = ii.track_id
GROUP BY t.name
ORDER BY purchase_count DESC
LIMIT 1;
"""

pd.read_sql(query, conn)
```

```python
## Q14 Average Price by Media Type 
query = """
SELECT 
    m.name AS media_type,
    AVG(t.unit_price) AS average_price
FROM tracks t
JOIN media_types m
    ON t.media_type_id = m.media_type_id
GROUP BY m.name
ORDER BY average_price DESC;
"""

pd.read_sql(query, conn)
```

```python
## Customer Analytics - Top Spending Customers 
query = """
SELECT 
    c.customer_id,
    c.first_name || ' ' || c.last_name AS customer_name,
    c.country,
    SUM(i.total) AS total_spent
FROM customers c
JOIN invoices i
    ON c.customer_id = i.customer_id
GROUP BY c.customer_id, customer_name, c.country
ORDER BY total_spent DESC
LIMIT 10;
"""

pd.read_sql(query, conn)
```

```python
## Average Customer Lifetime Value 
query = """
SELECT 
    AVG(customer_total) AS avg_customer_lifetime_value
FROM (
    SELECT 
        customer_id,
        SUM(total) AS customer_total
    FROM invoices
    GROUP BY customer_id
) sub;
"""

pd.read_sql(query, conn)
```

```python
## Repeat vs One-Time Customers 
query = """
SELECT 
CASE 
    WHEN purchase_count = 1 THEN 'One-Time Buyer'
    ELSE 'Repeat Customer'
END AS customer_type,
COUNT(*) AS total_customers
FROM (
    SELECT 
        customer_id,
        COUNT(invoice_id) AS purchase_count
    FROM invoices
    GROUP BY customer_id
) sub
GROUP BY customer_type;
"""

pd.read_sql(query, conn)
```

```python
## Revenue Per Customer by Country 
query = """
SELECT 
    c.country,
    SUM(i.total)/COUNT(DISTINCT c.customer_id) AS revenue_per_customer
FROM customers c
JOIN invoices i
    ON c.customer_id = i.customer_id
GROUP BY c.country
ORDER BY revenue_per_customer DESC;
"""

pd.read_sql(query, conn)
```

```python
## Inactive Customers (Last 6 Months) 
query = """
SELECT 
    c.customer_id,
    c.first_name || ' ' || c.last_name AS customer_name,
    MAX(i.invoice_date::DATE) AS last_purchase
FROM customers c
LEFT JOIN invoices i
    ON c.customer_id = i.customer_id
GROUP BY c.customer_id, customer_name
HAVING MAX(i.invoice_date::DATE) < (
    SELECT MAX(invoice_date::DATE) - INTERVAL '6 months'
    FROM invoices
);
"""

pd.read_sql(query, conn)
```

```python
## Section 2 — Sales & Revenue Analysis 
```

```python
## Sales Analysis - Monthly Revenue Trends 
query = """
SELECT 
    DATE_TRUNC('month', invoice_date::DATE) AS month,
    SUM(total) AS monthly_revenue
FROM invoices
WHERE invoice_date::DATE >= (
    SELECT MAX(invoice_date::DATE) - INTERVAL '2 years'
    FROM invoices
)
GROUP BY month
ORDER BY month;
"""

pd.read_sql(query, conn)
```

```python
# # What is the average value of an invoice (purchase)?
query = """
SELECT 
    AVG(total) AS average_invoice_value
FROM invoices;
"""

pd.read_sql(query, conn)
```

```python
## Which payment methods are used most frequently? 
query = """
SELECT 
    billing_country,
    COUNT(invoice_id) AS total_transactions
FROM invoices
GROUP BY billing_country
ORDER BY total_transactions DESC;
"""

pd.read_sql(query, conn)
```

```python
## How much revenue does each sales representative contribute? 
query = """
SELECT 
    e.employee_id,
    e.first_name || ' ' || e.last_name AS sales_rep,
    SUM(i.total) AS total_revenue
FROM employees e
JOIN customers c
    ON e.employee_id = c.support_rep_id
JOIN invoices i
    ON c.customer_id = i.customer_id
GROUP BY e.employee_id, sales_rep
ORDER BY total_revenue DESC;
"""

pd.read_sql(query, conn)
```

```python
### Which months or quarters have peak music sales? 
query = """
SELECT 
    DATE_TRUNC('month', invoice_date::DATE) AS month,
    SUM(total) AS total_revenue
FROM invoices
GROUP BY month
ORDER BY total_revenue DESC;
"""

pd.read_sql(query, conn)
```

```python
## Top Revenue Generating Tracks 
query = """
SELECT 
    t.track_id,
    t.name AS track_name,
    SUM(ii.unit_price * ii.quantity) AS total_revenue
FROM tracks t
JOIN invoice_items ii
    ON t.track_id = ii.track_id
GROUP BY t.track_id, t.name
ORDER BY total_revenue DESC
LIMIT 10;
"""

pd.read_sql(query, conn)
```

```python
## Which albums or playlists are most frequently included in purchases?  
query = """
SELECT 
    al.album_id,
    al.title AS album_name,
    COUNT(ii.invoice_line_id) AS purchase_count
FROM albums al
JOIN tracks t
    ON al.album_id = t.album_id
JOIN invoice_items ii
    ON t.track_id = ii.track_id
GROUP BY al.album_id, album_name
ORDER BY purchase_count DESC
LIMIT 10;
"""

pd.read_sql(query, conn)
```

```python
## Tracks Never Purchased
query = """
SELECT 
    t.track_id,
    t.name AS track_name
FROM tracks t
LEFT JOIN invoice_items ii
    ON t.track_id = ii.track_id
WHERE ii.invoice_line_id IS NULL;
"""

pd.read_sql(query, conn)
```

```python
## Average Track Price by Genre 
query = """
SELECT 
    g.genre_id,
    g.name AS genre_name,
    AVG(t.unit_price) AS avg_track_price
FROM genres g
JOIN tracks t
    ON g.genre_id = t.genre_id
GROUP BY g.genre_id, genre_name
ORDER BY avg_track_price DESC;
"""

pd.read_sql(query, conn)
```

```python
## Track Count vs Sales by Genre 
query = """
SELECT 
    g.genre_id,
    g.name AS genre_name,
    COUNT(DISTINCT t.track_id) AS total_tracks,
    COUNT(ii.invoice_line_id) AS total_sales
FROM genres g
LEFT JOIN tracks t
    ON g.genre_id = t.genre_id
LEFT JOIN invoice_items ii
    ON t.track_id = ii.track_id
GROUP BY g.genre_id, genre_name
ORDER BY total_sales DESC;
"""

pd.read_sql(query, conn)
```

```python
## Top 5 Highest Grossing Artists 
query = """
SELECT 
    ar.artist_id,
    ar.name AS artist_name,
    SUM(ii.unit_price * ii.quantity) AS total_revenue
FROM artists ar
JOIN albums al
    ON ar.artist_id = al.artist_id
JOIN tracks t
    ON al.album_id = t.album_id
JOIN invoice_items ii
    ON t.track_id = ii.track_id
GROUP BY ar.artist_id, artist_name
ORDER BY total_revenue DESC
LIMIT 5;
"""

pd.read_sql(query, conn)
```

```python
## Genre Popularity by Country 
query = """
SELECT 
    c.country,
    g.name AS genre_name,
    COUNT(ii.invoice_line_id) AS total_purchases
FROM customers c
JOIN invoices i
    ON c.customer_id = i.customer_id
JOIN invoice_items ii
    ON i.invoice_id = ii.invoice_id
JOIN tracks t
    ON ii.track_id = t.track_id
JOIN genres g
    ON t.genre_id = g.genre_id
GROUP BY c.country, genre_name
ORDER BY c.country, total_purchases DESC;
"""

pd.read_sql(query, conn)
```

```python
## Section 5 — Employee & Operational Efficiency 
```

```python
## Employees Managing Highest-Spending Customers 
query = """
SELECT 
    e.employee_id,
    e.first_name || ' ' || e.last_name AS employee_name,
    c.customer_id,
    c.first_name || ' ' || c.last_name AS customer_name,
    SUM(i.total) AS customer_total_spent
FROM employees e
JOIN customers c
    ON e.employee_id = c.support_rep_id
JOIN invoices i
    ON c.customer_id = i.customer_id
GROUP BY e.employee_id, employee_name, c.customer_id, customer_name
ORDER BY customer_total_spent DESC
LIMIT 10;
"""

pd.read_sql(query, conn)
```

```python
## Average Number of Customers per Employee 
query = """
SELECT 
    AVG(customer_count) AS avg_customers_per_employee
FROM (
    SELECT 
        support_rep_id,
        COUNT(customer_id) AS customer_count
    FROM customers
    GROUP BY support_rep_id
) sub;
"""

pd.read_sql(query, conn)
```

```python
## Revenue by Employee Region 
query = """
SELECT 
    e.employee_id,
    e.first_name || ' ' || e.last_name AS employee_name,
    c.country AS customer_region,
    SUM(i.total) AS total_revenue
FROM employees e
JOIN customers c
    ON e.employee_id = c.support_rep_id
JOIN invoices i
    ON c.customer_id = i.customer_id
GROUP BY e.employee_id, employee_name, customer_region
ORDER BY total_revenue DESC;
"""

pd.read_sql(query, conn)
```

```python
## Section 7 — Customer Retention & Purchase Patterns.
```

```python
## Purchase Frequency per Customer 
query = """
SELECT 
    c.customer_id,
    c.first_name || ' ' || c.last_name AS customer_name,
    COUNT(i.invoice_id) AS total_purchases
FROM customers c
JOIN invoices i
    ON c.customer_id = i.customer_id
GROUP BY c.customer_id, customer_name
ORDER BY total_purchases DESC;
"""

pd.read_sql(query, conn)
```

```python
## Average Time Between Customer Purchases 
query = """
SELECT 
    AVG(days_between) AS avg_days_between_purchases
FROM (
    SELECT 
        customer_id,
        invoice_date::DATE - 
        LAG(invoice_date::DATE) OVER (
            PARTITION BY customer_id 
            ORDER BY invoice_date::DATE
        ) AS days_between
    FROM invoices
) sub
WHERE days_between IS NOT NULL;
"""

pd.read_sql(query, conn)
```

```python
## Customers Buying Multiple Genres 
query = """
SELECT 
    ROUND(
        100.0 * COUNT(*) / (SELECT COUNT(*) FROM customers), 
        2
    ) AS percentage_multi_genre_customers
FROM (
    SELECT 
        c.customer_id
    FROM customers c
    JOIN invoices i
        ON c.customer_id = i.customer_id
    JOIN invoice_items ii
        ON i.invoice_id = ii.invoice_id
    JOIN tracks t
        ON ii.track_id = t.track_id
    GROUP BY c.customer_id
    HAVING COUNT(DISTINCT t.genre_id) > 1
) sub;
"""

pd.read_sql(query, conn)
```

```python
## Section 8 — Operational Optimization.
```

```python
## Most Common Track Combinations Purchased Together 
query = """
SELECT 
    t1.name AS track_1,
    t2.name AS track_2,
    COUNT(*) AS times_purchased_together
FROM invoice_items ii1
JOIN invoice_items ii2
    ON ii1.invoice_id = ii2.invoice_id
    AND ii1.track_id < ii2.track_id
JOIN tracks t1
    ON ii1.track_id = t1.track_id
JOIN tracks t2
    ON ii2.track_id = t2.track_id
GROUP BY track_1, track_2
ORDER BY times_purchased_together DESC
LIMIT 10;
"""

pd.read_sql(query, conn)
```

```python
## Pricing Patterns vs Sales 
query = """
SELECT 
    t.unit_price,
    COUNT(ii.invoice_line_id) AS total_sales,
    SUM(ii.unit_price * ii.quantity) AS total_revenue
FROM tracks t
JOIN invoice_items ii
    ON t.track_id = ii.track_id
GROUP BY t.unit_price
ORDER BY total_sales DESC;
"""

pd.read_sql(query, conn)
```

```python
## Media Type Usage Trends 
query = """
SELECT 
    m.name AS media_type,
    COUNT(ii.invoice_line_id) AS total_purchases
FROM media_types m
JOIN tracks t
    ON m.media_type_id = t.media_type_id
JOIN invoice_items ii
    ON t.track_id = ii.track_id
GROUP BY media_type
ORDER BY total_purchases DESC;
"""

pd.read_sql(query, conn)
```
